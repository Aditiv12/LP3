```python
df.drop_duplicates(inplace=True)
```

```python
df.fare_amount
```

```python
fare_amt = [fare for fare in df.fare_amount if fare >=0]
len(fare_amt)
```

```python
neg_fare_amt =[fare for fare in df.fare_amount if fare<0]
len(neg_fare_amt)
```

```python
df.fare_amount = [fare if fare>=0 else 0 for fare in df.fare_amount]
fare_amt = [fare for fare in df.fare_amount if fare ==0]
len(fare_amt)
```

```python
df['fare_amount'] = [df.fare_amount.mean() if fare==0 else fare for fare in df.fare_amount]
```

```python
fare_amt = [fare for fare in df.fare_amount if fare==0]
len(fare_amt)
```

```python
df = df[(df.pickup_latitude <90) & (df.dropoff_latitude >-90) & 
(df.pickup_latitude >-90) & (df.dropoff_latitude <90) & 
(df.pickup_longitude >-180) & (df.pickup_longitude <180) &
(df.dropoff_longitude >-180) & (df.dropoff_longitude <180)
]
```

```python
df['pickup_datetime'] =pd.to_datetime(df['pickup_datetime'])
```

```python
#feature engineering
df['year']=df['pickup_datetime'].dt.year
df['month']=df['pickup_datetime'].dt.month
df['day']=df['pickup_datetime'].dt.day
df['hour']=df['pickup_datetime'].dt.hour
```

```python
df.drop(['pickup_datetime'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['monthly_quarter']=df.month.map({1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4', 5: 'Q1', 6: 'Q2', 7: 'Q3', 8: 'Q4', 9: 'Q1', 10: 'Q2', 11: 'Q3', 12: 'Q4'})
df['hourly_segment']=df.hour.map({0: 'H1', 1: 'H1', 2: 'H1', 3: 'H1', 4: 'H2', 5: 'H2', 6: 'H2', 7: 'H2', 8: 'H3', 9: 'H3', 10: 'H3', 11: 'H3', 12: 'H4', 13: 'H4', 14: 'H4', 15: 'H4', 16: 'H5', 17: 'H5', 18: 'H5', 19: 'H5', 20: 'H6', 21: 'H6', 22: 'H6', 23: 'H6'})
```

```python
pip install geopy
```

```python
#distance calculation
from geopy.distance import geodesic
df['distance']= [round(geodesic((df.pickup_latitude[i], df.pickup_longitude[i]), (df.dropoff_latitude[i], df.dropoff_longitude[i])).km, 2) for i in df.index]
```

```python
df.drop(['month', 'hour'], axis=1, inplace=True)
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='distance', y='fare_amount')
plt.show()
```

```python
df = df[(df.fare_amount <250) & (df.fare_amount>0)]
df=df[df.distance <250]
```

```python
sns.scatterplot(data=df, x='distance', y='fare_amount')
plt.show()
```

```python
#remove outliers
df1 = df.copy()
features1 = ['fare_amount', 'distance']
for i in features1:
    Q1 =df1[i].quantile(0.25)
    Q3= df1[i].quantile(0.75)
    IQR= Q3-Q1
    df1 = df1[df1[i] <= (Q3+(1.5*IQR))]
    df1 = df1[df1[i] >= (Q1-(1.5*IQR))]
    df1 = df1.reset_index(drop=True)
```

```python
df1.head()
```

```python
df = df1.copy()
```

```python
sns.scatterplot(data=df, x='distance', y='fare_amount')
plt.show()
```

```python
df_dummy = pd.get_dummies(df[['monthly_quarter', 'hourly_segment']], drop_first=True)
```

```python
df=pd.concat([df, df_dummy], axis=1)
```

```python
df.head()
```

```python
df.drop(['monthly_quarter', 'hourly_segment'], axis=1, inplace=True)
```

```python
df.head()
```

```python
X=df.drop('fare_amount', axis=1)
y=df['fare_amount']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin
```

```python
lin.fit(X_train, y_train)
```

```python
y_pred = lin.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)*100
```
